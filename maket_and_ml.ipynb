# 마켓과 머신러닝

```python
import sklearn
```

```python
# 도미의 길이, 도미의 특성성

bream_length= [25.4,26.3,26.5,29.0,29.0,29.7,29,7,30.0,30.0,30.7,
            31.0,31.0,31.5,32.0,32.0,33.0,33.0,33.5,33.5,34.0,34.0,34.5,
            35.0,35.0,35.0,35.0,36.0,36.0,36.0,37.0,38.5,39.0,41.0,41.0]
bream_weight = [242.0,290.0,340.0,363.0,430.0,450.0,500.0,390.0,
                450.0,500.0,475.0,500.0,500.0,340.0,600.0,600.0,
                700.0,700.0,610.0,650.0,575.0,685.0,620.0,680.0,
                700.0,725.0,720.0,714.0,850.0,1000.0,920.0,955.0,
                925.0,975.0,950.0]
```

```python
# 산점도(scatter plot)

import matplotlib.pyplot as plt

plt.scatter(bream_length,bream_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
# 빙어 데이터

smelt_length = [9.8,10.5,10.6,11.0,11.2,11.3,11.8,11.8,12.0,12.2,
                12.4,13.0,14.3,15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2,13.4,
                12.2,19.7,19.9]

# 산점도
plt.scatter(bream_length,bream_weight)
plt.scatter(smelt_length,smelt_weight)
plt.xlabel('length') # x축 제목
plt.ylabel('weight') # y축 제목
plt.show()
```

# 도미와 빙어 데이터 합치기

```python
# 리스트 형태 데이터 합치기.
length = bream_length + smelt_length
weight = bream_weight + smelt_weight

# 사이킷 런을 사용하기위해 2차원 배열로 변형
fish_data = [[l,w] for l,w in zip(length,weight)]

# 정답 준비 (지도 학습)
# 1 : 도미, 0 : 빙어
fish_target = [1]*35 + [0]*14
```

# 머신러닝 알고리즘
## 1. k-최근접 이웃

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()

# 머신러닝 모델
# .fit(x_data , y_data)으로 학습
kn.fit(fish_data,fish_target)

# 입력 데이터 x에 대해 예측을 수행하고
# 그 결과가 y와 얼마나 맞는지를 평가
# score메소드 호출
# 정확도(accuracy) 0.0 ~ 1.0
kn.score(fish_data,fish_target)
```

```python
# -- 새로운 생선 예측 --

# predict()로 모델에 새로운 데이터를 예측
# 새로운 데이터는 2차원 데이터로 입력
kn. predict([[30,600]])
# -> 결과값 array([1]) : Numpy 배열 형태로 [1]을 반환
# 예측 결과가 1번 클래스(즉, 레이블 1에 해당)
```

```python
# 무조건 도미인 상황 만들기

# n_neighbors는 너무 작으면 과적합, 너무 크면 과소 적합 발생
# n_neighbors는 얼마나 많은 이웃을 참고할 것이지 결정하는 파라미터

kn49 = KNeighborsClassifier(n_neighbors=49)
kn49.fit(fish_data,fish_target)
kn49.score(fish_data,fish_target)
# n_neighbors=49로 전체 데이터를 참고하니 다수 클래스로만 예측되었음
# -> 도미만 나와서 35/49의 정확도가 나옴
print(35/49)
```
